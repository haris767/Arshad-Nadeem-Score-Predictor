```python
#task is to predict the score of a Arshad will obtain based on Throws we studied.


#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 #reading data
#read data in score predict variable (sp)
sp=pd.read_csv('/content/arshaddata.csv')
```

```python
#checking the data

sp.head()
```

```python
#exploring the relation between the data

sns.scatterplot(data=sp,x='speed k/h',y='throw')
```

```python
#th= sp['throw'].str.replace("m", "")

#print(th)
```

```python
 #defining data to be trained and tested

X = sp[['speed k/h']]
y= sp[['throw']]
```

```python
#importing library for spliting data

from sklearn.model_selection import train_test_split
```

```python
#spliting of data into train data and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
#importing Lin Reg model

from sklearn.linear_model import LinearRegression
```

```python
 #creating instance of the model

lm = LinearRegression()
```

```python
 #training the data

lm.fit(X_train,y_train)
```

```python
LinearRegression()
```

```python
#creating dataframe to represent the correlation

coeff_df = pd.DataFrame(lm.coef_,index=X.columns,columns=['Coeff'])
```

```python
 coeff_df

#this below dataframe means :
#keeping every other feature fixed, a unit increase in 'speed' feature will lead to 0.86 increase
```

```python
 #using the test data to predict

prediction = lm.predict(X_test)
```

```python
 #various prediction values

prediction
```

```python
 #original question for the task. If a Arshad speed more than previous speed then what score he can get?

speed = [[115.25]]
new_pred = lm.predict(speed)
```

**SVM MODEL USE FOR PREDICTION**

```python
from sklearn.svm import SVR
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [3,1,0.1,0.01]}
```

### **GridSearchCV a technique to search through the best parameter values from the given set of the grid of parameters**

```python
svr= GridSearchCV(SVR(epsilon=0.01,kernel='linear'),param_grid,verbose=3)
svr.fit(X_train,y_train)
```

```python
#print best parameter after tuning
print(svr.best_params_)
print(svr.best_estimator_)
```

Epsilon in SVM stands for epsilon insensitive loss (It denotes how much error you are willing to allow per training data instance)

```python
svr_best=SVR(epsilon=0.01,kernel='linear',C=1,gamma=3)
svr_best.fit(X_train,y_train)
```

```python
y_pred=svr_best.predict(X_test)
```

```python
#plotting the reggression line
line=svr_best.coef_*X+svr_best.intercept_
#plotting for the best data
plt.scatter(X,y,color='red')
plt.plot(X,line,color='blue')
plt.xlabel('speed')
plt.ylabel('throw')
plt.show()
```
